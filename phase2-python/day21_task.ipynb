```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("EDA_Data.csv")

plt.figure(figsize=(8,5))

plt.scatter(
    df["Quantity"],
    df["Sales"],
    color="purple",
    s=100
)

plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["Sales"],
    bins=10,
    color="orange"
)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```
