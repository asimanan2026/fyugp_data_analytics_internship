```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("sales_data.csv")
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(6,4))

plt.bar(category_sales.index, category_sales.values, color="black")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.plot(df["Sales"], marker="o",color="Black")

plt.title("Sales Line Chart")
plt.xlabel("Index")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.hist(df["Sales"], bins=10,color="Black")

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```
