```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("day24_data.csv")
```

```python
df.isnull().sum()
```

```python
df["Profit"] = df["Profit"].fillna(df["Profit"].median())
```

```python
df["Rating"] = df["Rating"].fillna(df["Rating"].mode()[0])
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Customer_Age']=pd.to_numeric(df['Customer_Age'],errors='coerce')
```

```python
df["Customer_Age"] = df["Customer_Age"].fillna(df["Customer_Age"].mean())
```

```python
df.info()
```

```python
df["Rating"] = df["Rating"].replace("Excellent",5)
```

```python
df["Rating"] = pd.to_numeric(df["Rating"])
```

```python
df[df["Sales"] < 0]
```

```python
df = df[df["Sales"] >= 0]
```

```python
df.info()
```

```python
df.describe()
```

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["Sales"] >= lower) & (df["Sales"] <= upper)]
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    category_sales.index,
    category_sales.values,
    color="black"
)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Quantity"],
    df["Sales"],
    color="black",
    s=100
)

plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["Sales"],
    bins=10,
    color="black"
)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```
