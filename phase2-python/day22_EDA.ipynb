```python
df.groupby("Category")["Sales"].sum()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))

plt.bar(
    category_sales.index,
    category_sales.values,
    color="#19222A"
)

plt.title("Total Sales by Category", fontsize=16, fontweight="bold")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Quantity"],
    df["Sales"],
    color="#D2212D"
)

plt.title("Quantity vs Sales", fontsize=16, fontweight="bold")
plt.xlabel("Quantity")
plt.ylabel("Sales")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["Sales"],
    bins=10,
    color="#13422D"
)

plt.title("Distribution of Sales", fontsize=16, fontweight="bold")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```
