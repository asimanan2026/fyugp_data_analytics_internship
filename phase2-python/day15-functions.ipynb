```python
def celsius_to_farenheit(c):
    f=(c*9/5)+32
    return f

print(celsius_to_farenheit(0))
print(celsius_to_farenheit(32))
print(celsius_to_farenheit(74))
```

```python
def get_grade(score):
    if(score>=90):
        return "Grade A"
    elif(score>=75):
        return "Grade B"
    elif(score>=50):
        return "Grade C"
    else:
        return "Fail"

print(get_grade(98))
print(get_grade(79))
print(get_grade(52))
print(get_grade(29))
print(get_grade(2))
```
