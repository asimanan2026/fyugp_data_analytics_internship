```python
scores = [45, 67, 89, 32, 76, 50, 91, 28, 60, 73]
for score in scores:
    if score >= 50:
        print(score, "- Pass")
    else:
        print(score, "- Fail")
```

```python
pass_count=0
fail_count=0
for score in scores:
    if score>=50:
        pass_count+=1
    else:
        fail_count+=1

print("Students Passed",pass_count)
print("Students Failed",fail_count)
```

```python
print(pass_count,"Students passed out of",len(scores))
```
